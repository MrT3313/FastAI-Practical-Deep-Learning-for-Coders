# Is it a bird?

```python
from fastbook import *
from pprint import pprint
import random
from PIL import Image as PILImage
from IPython.display import Image, display
```

```python
from parseimagenet import get_image_paths_by_keywords
from parseimagenet import get_available_presets, KEYWORD_PRESETS
!pip show parseimagenet
```

```python
# Set the path to your ImageNet directory
base_path = Path('/Users/mrt/Documents/MrT/code/computer-vision/image-bank/ImageNet-Subset')
```

## Get Images

```python
number_of_images = 400
bird_image_paths = get_image_paths_by_keywords(
    base_path=base_path,
    preset="birds",
    num_images=number_of_images
)
print(f'{len(bird_image_paths)} Bird Images')
pprint(bird_image_paths[:2])

print("")

dog_image_paths = get_image_paths_by_keywords(
    base_path=base_path,
    preset="dogs",
    num_images=number_of_images
)
print(f'{len(dog_image_paths)} Dog Images')
pprint(dog_image_paths[:2])

print("")

snake_image_paths = get_image_paths_by_keywords(
    base_path=base_path,
    preset="snakes",
    num_images=1
)
print(f'{len(snake_image_paths)} Snake Images')
pprint(snake_image_paths[0])
```

```python
random_index = random.randrange(len(bird_image_paths))
display(Image(filename=bird_image_paths[random_index]))

random_index = random.randrange(len(dog_image_paths))
display(Image(filename=dog_image_paths[random_index]))
```

# Create Data Block

```python
# CLASSIFICATION HACK: https://github.com/MrT3313/Parse-ImageNet/issues/3
def get_label(path):
    if path in bird_image_paths:
        return "bird"
    else:
        return "dog"

# Convert to sets for faster lookup
bird_set = set(bird_image_paths)

def get_label(path):
    return "bird" if path in bird_set else "dog"

# DATA BLOCK
dls = DataBlock(
    blocks=(ImageBlock, CategoryBlock),
    splitter=RandomSplitter(valid_pct=0.2, seed=42),
    # get_y=parent_label,
    get_y=get_label,
    item_tfms=[Resize(192, method='squish')]
).dataloaders(bird_image_paths + dog_image_paths)

dls.show_batch(max_n=6)
```

# Train Model

```python
learn = vision_learner(dls, resnet18, metrics=error_rate)
learn.remove_cb(ProgressCallback) # getting errors w/ progress output
learn.fine_tune(3)

# epoch | train_loss | valid_loss | error_rate | time
```

# Predict

```python
# MODEL TRAINED ON DOGS & BIRDS
is_bird, _bird, probs_bird_input = learn.predict(bird_image_paths[random_index])
print(f'This is a: {is_bird}')
print(f"Probability it's a bird: {probs_bird_input[0]:.4f}")
pprint(_bird)
pprint(probs_bird_input)

print('')

is_dog, _dog, probs_dog_input = learn.predict(dog_image_paths[random_index])
print(f'This is a: {is_dog}')
print(f"Probability it's a bird: {probs_dog_input[0]:.4f}")
pprint(_dog)
pprint(probs_dog_input)

print('')

# MODEL HAS NO IDEA WHAT A SNAKE IS!
is_snake, _snake, probs_snake_input = learn.predict(snake_image_paths[0])
print(f'This is a: {is_snake}')
print(f"Probability it's a bird: {probs_snake_input[0]:.4f}")
pprint(_snake)
pprint(probs_snake_input)
```
